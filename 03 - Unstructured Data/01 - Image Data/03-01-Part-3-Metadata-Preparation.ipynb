# Mendapatkan semua image path

```python
from glob import glob
```

```python
fname = glob("/kaggle/input/number-image/data/*/*.png") # kita ganti nama folder dan file dengan bintang (*) dan glob akan otomatis men-detect
fname[:10]
```

# X = Path -> Image -> Flatten

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.array([plt.imread(f).flatten() for f in fname])
X
```

```python
X.shape
```
