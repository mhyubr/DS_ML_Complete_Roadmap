```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from jcopml.pipeline import num_pipe, cat_pipe
from jcopml.utils import save_model, load_model
from jcopml.plot import plot_missing_value
from jcopml.feature_importance import mean_score_decrease
```

# Import Data

```python
df = pd.read_csv('/kaggle/input/small-mnist/small_mnist.csv')
df.head()
```

# Dataset Splitting

```python
df.iloc[1:].values.max() 
```

`Maksimal data selalu 255`. sehingga ketika di scaling harus `dibagi 255` 

```python
X = df.drop(columns='label').values / 255 # untuk gambar tidak menggunakan DataFrame Pandas tetapi menggunakan Array Numpy
y = df.label.values # untuk gambar tidak menggunakan DataFrame Pandas tetapi menggunakan Array Numpy

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, X_train.shape, y_train.shape
```

# Visualize Random Data

```python
import matplotlib.pyplot as plt
```

```python
N = len(X_train)
idx = np.random.choice(N)
img, label = X_train[idx], y_train[idx]

plt.imshow(img.reshape(28, 28), cmap='gray')
plt.title(label)
plt.axis('off');
```

# Training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from jcopml.tuning import random_search_params as rsp
```

```python
rf_pipeline = Pipeline([
    ('algo', RandomForestClassifier(n_jobs=-1, random_state=42))
])

rf_model = RandomizedSearchCV(rf_pipeline, rsp.rf_params, n_iter=50, cv=3, n_jobs=-1, verbose=1, random_state=42)
rf_model.fit(X_train, y_train)

print(rf_model.best_params_)
print(rf_model.score(X_train, y_train), rf_model.best_score_, rf_model.score(X_test, y_test))
```

# Visualize Prediction

```python
idx = np.random.choice(len(X_test), size=36, replace=False)
images, labels = X_test[idx], y_test[idx]
preds = rf_model.predict(images)

fig, axes = plt.subplots(6, 6, figsize=(15, 15))
for img, label, pred, ax in zip(images, labels, preds, axes.flatten()):
    font = {"color": "g"} if label == pred else {"color": "r"}
    
    ax.imshow(img.reshape(28, 28), cmap="gray")
    ax.set_title(f"Label: {label} | Pred: {pred}", fontdict=font)
    ax.axis('off');
```
