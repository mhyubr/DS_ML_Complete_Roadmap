```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 10)
y = np.sqrt(X)
y1 = np.sin(X)
y2 = np.cos(X)
```

# Inisialisasi Figure dan Axes

```python
fig = plt.figure()
ax = plt.axes()
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y1)
ax.plot(X, y2)
```

## Shortcut

```python
plt.figure(figsize=(6, 6))
plt.plot(X, y1, 'r--')
plt.plot(X, y2, 'gx')
```

```python
plt.figure(figsize=(6, 6))
plt.plot(X, y1, 'r--')
plt.plot(X, y2, 'gx')
plt.xlim(0, 20)
plt.ylim(-2, 2)
```

```python
plt.figure(figsize=(6, 6))
plt.plot(X, y1, 'r--', linewidth=5)
plt.plot(X, y2, 'gx', markersize=10.5)
plt.xlim(0, 20)
plt.ylim(-2, 2)
```

```python
plt.figure(figsize=(6, 6))
plt.plot(X, y1, 'r--', linewidth=5)
plt.plot(X, y2, 'gx', markersize=10.5)
plt.xlim(0, 20)
plt.ylim(-2, 2)
plt.axis('equal')
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y1, 'r--', linewidth=5)
ax.plot(X, y2, 'gx', markersize=10.5)
ax.set(xlim=(0, 20), ylim=(-2, 2))
ax.axis('equal')
```

```python
plt.figure(figsize=(6, 6))
plt.plot(X, y1, 'r--', linewidth=5)
plt.plot(X, y2, 'gx', markersize=10.5)
plt.xlim(0, 20)
plt.ylim(-2, 2)
plt.axis('equal')
plt.title('Sin dan Cos')
plt.xlabel('X')
plt.ylabel('y')
```

```python
plt.figure(figsize=(6, 6))
plt.plot(X, y1, 'r--', linewidth=5, label='sin')
plt.plot(X, y2, 'gx', markersize=10.5, label='cos')
plt.xlim(0, 20)
plt.ylim(-2, 2)
plt.axis('equal')
plt.title('Sin dan Cos', fontsize=14, color='b')
plt.xlabel('X', fontsize=11, color='r')
plt.ylabel('y', fontsize=11, color='g')
plt.legend(loc='best', ncol=2, shadow=True, borderpad=1)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = plt.axes()
ax.plot(X, y1, 'r--', linewidth=5, label='sin')
ax.plot(X, y2, 'gx', markersize=10.5, label='cos')
ax.set(xlim=(0, 20), ylim=(-2, 2), title='Sin dan Cos', xlabel='x', ylabel='y')
ax.axis('equal')
ax.legend(loc='lower center', ncol=2, shadow=True, borderpad=1)
```

```python
df = pd.read_csv('/kaggle/input/circle-data/circle_data.csv', delimiter=';')
df.head()
```

```python
X = df.x
y = df.y
rad = df.radius
```

```python
plt.scatter(X, y, c='r', s=100)
plt.axis('equal');
```

```python
plt.scatter(X, y, c=rad, s=rad*100, cmap='seismic')
plt.axis('equal');
```

# Histogram

```python
numbers = np.random.randn(10000)
plt.hist(numbers, bins=80, density=1);
```

# Subplot

### Matlab Style

```python
plt.figure(figsize=(15, 10))

plt.subplot(231)
plt.hist(numbers, bins=80, density=1);

plt.subplot(233)
plt.hist(numbers, bins=80, density=1, cumulative=1);

plt.subplot(235)
plt.scatter(X, y, c=rad, s=rad*100, cmap='seismic')
plt.axis('equal');

plt.subplot(236)
plt.hist(numbers, bins=20, density=1);
```

### Matplotlib Style

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 10))

ax[0,0].hist(numbers, bins=80, density=1);
ax[0,2].hist(numbers, bins=80, density=1, cumulative=1);
ax[1,1].scatter(X, y, c=rad, s=rad*100, cmap='seismic')
ax[1,1].axis('equal');
ax[0,2].hist(numbers, bins=20, density=1);
```

# Text dan Annotation

```python
numbers = np.random.randn(10000)
```

```python
plt.figure(figsize=(6,6))
plt.hist(numbers, bins=80, density=1);
plt.ylim(0, 1)
plt.axvline(0, color='r')
plt.text(0, 0.8, "mean=0")
```

```python
plt.figure(figsize=(6,6))
plt.hist(numbers, bins=80, density=1);
plt.ylim(0, 1)
plt.axvline(0, color='r')
plt.text(0, 0.8, "mean=0", ha='center')
```

```python
plt.figure(figsize=(6,6))
plt.hist(numbers, bins=80, density=1);
plt.ylim(0, 1)
plt.axvline(0, color='r')
plt.text(0, 0.8, "mean=0", ha='center')
plt.annotate("mean=0", xy=(0, 0.4), xytext=(1, 0.5), ha='center', arrowprops={'arrowstyle': '-|>'})
```

# 3D Plot

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.Figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
```

```python
x = np.linspace(-6, 6, 5)
y = np.linspace(-6, 6, 5)
```

```python
X, y = np.meshgrid(x, y)
```

```python
X
```

```python
y
```

```python
z = np.sqrt(X**2 + y**2 + 10)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(X, y, z)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(X, y, z, antialiased=False, cmap='winter')
ax.set(xlabel='x', ylabel='y', zlabel='z')
```

```python
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(X, y, z, antialiased=False, cmap='winter')
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=10, azim=60)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, y, z, cmap='winter')
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=10, azim=60)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
ax.contour3D(X, y, z, cmap='winter')
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=10, azim=60)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
ax.contour3D(X, y, z, 50, cmap='winter')
ax.set(xlabel='x', ylabel='y', zlabel='z')
ax.view_init(elev=10, azim=60)
```
