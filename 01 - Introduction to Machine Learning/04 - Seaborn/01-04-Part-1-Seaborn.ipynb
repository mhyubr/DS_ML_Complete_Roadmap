# Seaborn & Pandas DataFrame

```python
df = pd.read_csv('/kaggle/input/cerealcsv/01 - Introduction to Machine Learning/02 - Pandas Introduction/dataset/cereal.csv', index_col='name')
df.head()
```

## Missing Value

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.isna(), cmap='Blues')
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(~(df.isna()), cmap='Blues', cbar=False)
```

## Correlation

```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    df.corr(),
    cmap='bwr',
    vmin=-1,
    vmax=1,
    annot=True,
    square=True,
    fmt='.2f',
    annot_kws={'size': 8}
)
```

## Pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='shelf', vars=['calories', 'fat', 'rating'])
```

```python
sns.pairplot(df, hue='shelf', vars=['calories', 'fat', 'rating'], kind='reg')
```

## Jointplot

```python
sns.jointplot(data=df, x='potass', y='fiber', marginal_kws={'bins': 20}, kind='reg')
```

## Categorical Data

### Distribution

```python
sns.catplot(data=df,  x='mfr', y='rating')
```

```python
sns.boxplot(data=df,  x='mfr', y='rating')
```

```python
sns.violinplot(data=df,  x='mfr', y='rating')
```

### Count

```python
sns.countplot(data=df, x='mfr')
```

```python
sns.countplot(data=df, x='mfr', hue='shelf')
```
