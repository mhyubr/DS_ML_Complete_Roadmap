# Conditionals

```python
score = 50 
```

```python
if score < 0:
    print('Nilai Kerendahan!')
elif score < 60:
    print('Indeks: C')
elif score < 80:
    print('Indeks: B')
elif score < 100:
    print('Indeks: A')
elif score == 100:
    print('Indeks: S')
else:
    print('Score ketinggian!')
```

```python
def konversi_ke_indeks(score):
    if score < 0:
        return 'Nilai Kerendahan!'
    elif score < 60:
        return 'Indeks: C'
    elif score < 80:
        return 'Indeks: B'
    elif score < 100:
        return 'Indeks: A'
    elif score == 100:
        return 'Indeks: S'
    else:
        return 'Score ketinggian!'
```

```python
konversi_ke_indeks(-100)
```
