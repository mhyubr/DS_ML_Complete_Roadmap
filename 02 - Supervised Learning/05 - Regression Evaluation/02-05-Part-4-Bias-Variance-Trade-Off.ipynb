## The Principle of bias-variance tradeoff
More Complex Model -> `Low bias, High variance` -> Prone to `Overfit`

More Simple Model -> `High bias, Low variance` -> Prone to `Underfit` 

```python
illustration.bias_variance_tradeoff
```

```python
illustration.just_right
```

## Contoh Analisis
`Linear Regression` -> less feature -> less complex -> prone to `Underfit`

`Poly Regression` -> more feature -> more complex -> prone to `Overfit`

# Using bias-variance trade off in tuning

### KNN - n_neighbor

```python
demo.knn()
```

Pada KNN `semakin banyak` jumlah tetangga (n_neighbor) maka `semakin simple` modelnya (underfit)

### SVM (Kernel)

```python
demo.svm_kernel()
```

Pada SVM menggunakan `Kernel = RBF` cenderung `Overfit`, sedangkan menggunakan `Kernel = Linear` lebih cenderung `Underfit`

### SVM - C(Penalty)

```python
demo.svm_c()
```

Pada SVM `semakin banyak` jumlah toleran (penalty) maka `semakin kompleks` modelnya (overfit)

### SVM - Gamma (detailness)

```python
demo.svm_gamma()
```

Pada SVM `semakin detail oriented` maka `semakin kompleks` modelnya (overfit)
