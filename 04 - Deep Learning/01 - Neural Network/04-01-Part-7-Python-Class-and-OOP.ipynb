```python
class Manusia:
    def __init__(self, warna_rambut, tinggi=150, berat=60):
        self.warna_rambut = warna_rambut
        self.tinggi = tinggi
        self.berat = berat
        
    def cat_rambutnya(self, warna):
        self.warna_rambut = warna
        print(f"Rambutnya jadi warna {warna}")
        
    def BMI(self):
        return self.berat / self.tinggi ** 2
```

```python
Asep = Manusia("hitam", 178, 60)
```

```python
Asep.cat_rambutnya("merah")
```

# Inheritance

```python
class Mahasiswa(Manusia):
    def __init__(self, warna_rambut, tinggi, berat, univ):
        super().__init__(warna_rambut, tinggi, berat) # mengambil sifat dari Manusia
        self.universitas = univ
        
    def pindah_univ(self, univ_baru):
        self.universitas = univ_baru
```

```python
otong = Mahasiswa("hitam", 150, 45, "TelU")
```

```python
otong.universitas
```

```python
otong.pindah_univ("ITB")
otong.universitas
```

```python
class CustomNN(nn.Module):
    def __init__(self, input_size, output_size):
        super().__init__()
        self.fc = nn.Sequential( # fc = fully connected / Dense
            nn.Linear(input_size, 16),
            nn.ReLU(),
            nn.Linear(16, 8),
            nn.ReLU(),
            nn.Linear(8, output_size),
            nn.Sigmoid()
        )
        
    def forward(self, x): # syarat menggunakan nn.Module harus mendefinisikan feedforward
        return self.fc(x)
```

## Multitask Learning (Klasifikasi + Regresi)

```python
class FaceBiometric(nn.Module):
    def __init__(self, input_size, output_size):
        super().__init__()
        self.fc = nn.Sequential(
            nn.Linear(input_size, 16),
            nn.ReLU(),
        )
        
        self.reg = nn.Linear(16, 4)
        self.clf = nn.Sequential(
            nn.Linear(16, 4),
            nn.Sigmoid()
        )
        
    def forward(self, x):
        x = self.fc(x)
        pred_reg =  self.reg(x)
        pred_clf =  self.clf(x)
        return pred_reg, pred_clf
```
